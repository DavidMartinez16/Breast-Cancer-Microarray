# Importar las librerías

```python
import pandas as pd
import mglearn
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

# Importar el conjunto de datos

```python
# Leer el conjunto de datos procesado anteriormente
data=pd.read_csv('Breast_GSE45827.csv')
data.head()
```

```python
# Elimino las columnas inncesarias para el EDA

# Se elimina la columna de muestras ya que no aporta información relevante
datac=data.drop(['samples'],axis=1)

# Cambio los datos del tipo de cáncer en datos numéricos
datac['type'].replace(['basal','HER','cell_line','normal','luminal_A','luminal_B'],[1,2,3,4,5,6],inplace=True)
```

# Visualización de la distribución de cada una de las clase

Para visualizar la dispersión de cada uno de los 6 tipos de cáncer de mama, se utilizarán técnicas de reducción de la dimensionalidad con el objetivo de graficar en 2 dimensiones dicha dispersión

# Análisis de Componentes Principales (PCA)

```python
#  -------------- ANÁLSIS DE COMPONENTES PRINCIPALES - PCA -------------------
plt.figure(figsize=(12,7))
pca=PCA(n_components=2) # Número de componentes a transformar
data_transform=pca.fit_transform(datac)
# Verifico la forma de los datos antes y despues de la transformación
print('Shape del antiguo data : ', datac.shape)
print('Shape del data transformado : ', data_transform.shape)

# Grafico el nuevo conjunto de datos con PCA
figurepca=mglearn.discrete_scatter(data_transform[:,0], data_transform[:,1],data['type'])
plt.legend(loc='center left' , bbox_to_anchor=(1, 0.5))
plt.title('Dispersión de los datos con PCA', weight='bold',fontsize=20)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.grid()
plt.show(figurepca)
```

# Incrustación de Vecinos Estocásticos (TSNE)

```python
plt.figure(figsize=(12,7))
tsne=TSNE(n_components=2) # Número de componentes a transformar
data_tsne=tsne.fit_transform(datac)
# Verifico la forma de los datos antes y despues de la transformación
print('Shape del antiguo data : ', datac.shape)
print('Shape del data transformado : ', data_tsne.shape)

# Grafico el nuevo conjunto de datos con TSNE
figuretsne=mglearn.discrete_scatter(data_tsne[:,0], data_tsne[:,1],data['type'])
plt.legend(loc='center left' , bbox_to_anchor=(1, 0.5))
plt.title('Dispersión de los datos con TSNE', weight='bold',fontsize=20)
plt.xlabel('TSNE 1')
plt.ylabel('TSNE 2')
plt.grid()
plt.show(figuretsne)
```

# Número de instancias por clase

```python
plt.figure(figsize=(12,7))
sns.countplot(data=data,x='type')
plt.title('Número de Instancias por Clase', weight='bold',fontsize=20)
plt.plot()
```
